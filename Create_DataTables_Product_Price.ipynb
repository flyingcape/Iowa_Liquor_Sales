# Clean & Create Product Price Table

```python
product_price_col = ['ItemNumber',
                  'Date',
                  'StateBottleRetail',
                  'StateBottleCost']

product_price = sales[product_price_col].copy()
```

## ðŸ§¹ Data Cleaning - Product Price Table 

```python
# Rename some columns
product_price.rename(columns = {'ItemNumber':'ProductID'}, inplace = True)
```

```python
product_price['Date'] = pd.to_datetime(product_price['Date'])
product_price['ProductID'] = product_price['ProductID'].apply(str)
```

```python
# will use these to fill missing data
product_price['Year'] = pd.to_datetime(product_price['Date']).dt.year
product_price['Month'] = pd.to_datetime(product_price['Date']).dt.month
```

```python
product_price.head()
```

### I. Fill-in Missing Price Data

```python
product_price.isnull().sum(axis = 0)
```

```python
product_price.loc[product_price['StateBottleRetail'].isnull()]
```

```python
print('the product numbers are: ',product_price.loc[product_price['StateBottleRetail'].isnull(),'ProductID'].unique())
```

```python
# Checking ou this particular product ID price on Feb 2012 
product_price.loc[(product_price['ProductID']=='38180') & (product_price['Year']==2012) & (product_price['Month']==2)]
```

```python
# Checking ou this particular product ID price on Feb 2012 
product_price.loc[(product_price['ProductID']=='35920') & (product_price['Year']==2012) & (product_price['Month']==2)]
```

```python
# Fill Missing (end of Jan 2012) with data from February 2012 - doing this MANUALLY
product_price.loc[(product_price['ProductID']=='38180') & (product_price['StateBottleRetail'].isnull()),'StateBottleRetail'] = 15.05
product_price.loc[(product_price['ProductID']=='38180') & (product_price['StateBottleCost'].isnull()),'StateBottleCost'] = 10.03

product_price.loc[(product_price['ProductID']=='35920') & (product_price['StateBottleRetail'].isnull()),'StateBottleRetail'] = 7.5
product_price.loc[(product_price['ProductID']=='35920') & (product_price['StateBottleCost'].isnull()),'StateBottleCost'] = 5.0
```

```python
# Checking the result
product_price.loc[((product_price['ProductID']=='38180')|(product_price['ProductID']=='35920')) & (product_price['Year']==2012) & (product_price['Month']==1)]
```

### Before Moving On: Consolidate ProductIDs for same date

```python
product_price.groupby(['ProductID','Date']).size().sort_values()
```

```python
# Take average of Prices for a given day if there are multiple
product_price = product_price.groupby(['ProductID', 'Date']).agg({'StateBottleRetail':'mean', 'StateBottleCost':'mean'})\
                .reset_index()
```

```python
product_price.groupby(['ProductID','Date']).size().sort_values()
```

```python
# Drop duplicates
product_price.drop_duplicates(keep = 'first', inplace = True)
product_price.shape
```
