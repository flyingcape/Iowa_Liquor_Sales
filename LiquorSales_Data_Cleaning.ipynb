# 2. Product Table

```python
product_col = ['ItemNumber',
            'ItemDescription',
            'Category',
            'CategoryName',
            'Pack',
            'BottleVolume(ml)',
            'VendorNumber',
            'Date',
            'Year']

product = sales[product_col].copy()
```

## ðŸ§¹ Data Cleaning - Product Table

```python
# Rename some columns
product.rename(columns = {'ItemNumber':'ProductID',
                          'ItemDescription':'ProductDescription',
                          'Category':'CategoryCode',
                          'BottleVolume(ml)':'BottleVolume',
                          'VendorNumber':'VendorID'}, inplace = True)
# Make lower-case
product['CategoryName'] = product['CategoryName'].str.lower()
product['ProductDescription'] = product['ProductDescription'].str.lower()
```

```python
product.head()
```

```python
product.info()
```

```python
product.shape[0]
```

### I. Looking at ProductID numbers with missing CategoryCode data

```python
# Converting all of these to strings, some are numeric, others are strings
# cannot convert to numeric because I get an error for a combo ID, e.g. x1000 (not a real ID)
product['ProductID'] = product['ProductID'].apply(str)
```

```python
# Which ProductID numbers are missing CategoryCode data?
ProductIDs_missing_category = product.loc[product['CategoryCode'].isnull(),'ProductID'].unique()
ProductIDs_missing_category
```

```python
%%time
# if CategoryCode information is available on other rows, use the last CategoryCode & Category Name to fill in the information
for ID in ProductIDs_missing_category:
    # print(ID)
    temp = product.loc[product['ProductID'].isin([ID])] # create a smaller table of this product id
    temp = temp.sort_values(['Date'])
    ncodes_with_data = temp['CategoryCode'].nunique() 
    # print(ncodes_with_data)
    temp_notnull = temp.loc[temp['CategoryCode'].notnull()]

    # print(temp_notnull)
    if ncodes_with_data > 0:
        product.loc[((product['ProductID'].isin([ID]))&(product['CategoryCode'].isnull())),'CategoryName'] = temp_notnull['CategoryName'][temp_notnull.index[-1]]
        product.loc[((product['ProductID'].isin([ID]))&(product['CategoryCode'].isnull())),'CategoryCode'] = temp_notnull['CategoryCode'][temp_notnull.index[-1]]
```

```python
# The remaining ProductID numbers that are still missing CategoryCode data
product.loc[product['CategoryCode'].isnull(),'ProductID'].unique()
```

### II. Looking at CategoryCode numbers with mulitiple CategoryName values

```python
print('Number of Category Codes: ', product['CategoryCode'].nunique())
```

```python
product.groupby(['CategoryCode','CategoryName'])['Pack'].count()
```

```python
# looking at an example
test = product.sort_values(['CategoryCode','CategoryName','Date'])
test.loc[test['CategoryCode']==1022200]
```

```python
# Create a CategoryCode list with the most recent category names
CategoryCodes_df = product.loc[(product['CategoryName'].notnull())].sort_values(['CategoryCode','CategoryName','Date']).drop_duplicates(subset = ['CategoryCode'], keep = 'last')
CategoryCodes_df.head(25)
```

```python
CategoryCodes_df.shape[0]
```

```python
# Consolidate the CategoryName values
Merged_Product_Category = product.merge(CategoryCodes_df[['CategoryCode', 'CategoryName']], on = 'CategoryCode', how = 'left')
Merged_Product_Category.drop(columns = 'CategoryName_x', inplace = True)
Merged_Product_Category.rename(columns = {'CategoryName_y': 'CategoryName'}, inplace = True)
product = Merged_Product_Category.copy()
product.head()
```

```python
# Checking that there is only one CategoryName per CategoryCode
product.groupby(['CategoryCode','CategoryName'])['Pack'].count()
```

### III. Looking at ProductID numbers with mulitiple ProductDescription entries

```python
# Create a ProductID list with the most recent category names
ProductID_df = product.loc[(product['ProductDescription'].notnull())].sort_values(['ProductID','ProductDescription','Date']).drop_duplicates(subset = ['ProductID'], keep = 'last')
ProductID_df.head(25)
```

```python
# Consolidate the ProductDescription values
Merged_Product_Descrip = product.merge(ProductID_df[['ProductID', 'ProductDescription']], on = 'ProductID', how = 'left')
Merged_Product_Descrip.drop(columns = 'ProductDescription_x', inplace = True)
Merged_Product_Descrip.rename(columns = {'ProductDescription_y': 'ProductDescription'}, inplace = True)
product = Merged_Product_Descrip.copy()
product.head()
```

### IV. Add Proof data

```python
proof = product_df[['ItemNumber', 'Proof']]
```

```python
proof.head()
```

```python
proof.rename(columns = {'ItemNumber':'ProductID'}, inplace = True)
proof['ProductID'] = proof['ProductID'].apply(str)
```

```python
# Attaching proof to the product database.
Proof_Merge = product.merge(proof[['ProductID', 'Proof']], on = 'ProductID', how = 'left')
Proof_Merge.head()
```

### V. Add Category Groups

```python
product = Proof_Merge.copy()
```

```python
product['CategoryName'] = product['CategoryName'].apply(str)
# Create category grouping.
product['CategoryGroup'] = np.where(product['CategoryName'].str.contains('rum'),'rum',
                                    np.where(product['CategoryName'].str.contains('gin'),'gin', 
                                    np.where(product['CategoryName'].str.contains('cocktail'), 'cocktail',
                                    np.where(product['CategoryName'].str.contains('brandy|brandies'), 'brandy',
                                    np.where(product['CategoryName'].str.contains('whisk|bourbon|scotch'), 'whisky',
                                    np.where(product['CategoryName'].str.contains('tequila|mezcal'), 'tequila',
                                    np.where(product['CategoryName'].str.contains('schnapps'), 'schnapps',
                                    np.where(product['CategoryName'].str.contains('vodka'), 'vodka',
                                    np.where(product['CategoryName'].str.contains('spirit'), 'spirit',
                                    np.where(product['CategoryName'].str.contains('liqueur'), 'liqueur',
                                    np.where(product['CategoryName'].str.contains('brandies'), 'brandy', 'other')))))))))))
```

# 2. Product Price Table

```python
product_price_col = ['ItemNumber',
                  'Date',
                  'StateBottleRetail',
                  'StateBottleCost']

product_price = sales[product_price_col].copy()
```

## ðŸ§¹ Data Cleaning - Product Price Table 

```python
# Rename some columns
product_price.rename(columns = {'ItemNumber':'ProductID'}, inplace = True)
```

```python
product_price['Date'] = pd.to_datetime(product_price['Date'])
```

```python
product_price.head()
```

```python
# if inconsistencies exist for a ProductID (for a given day), use the mean of StateBottleRetail and StateBottleCost respectively
product_price = product_price.groupby(['ProductID', 'Date']).agg({'StateBottleRetail':'mean', 'StateBottleCost':'mean'}).reset_index()
```

## âœ… Final Data Cleaning - Product Price Table

```python
product_price.shape
```

```python
# Drop duplicates
product_price.drop_duplicates(keep = 'first', inplace = True)
product_price.shape
```

```python
product_price.head()
```

```python
product_price.isnull().sum(axis = 0)
```

```python
# Removing missing
product_price.dropna(inplace = True)
product_price.isnull().sum(axis = 0)
```
