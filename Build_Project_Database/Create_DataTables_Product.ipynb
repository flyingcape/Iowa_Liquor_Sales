# Clean & Create Product Table

```python
product_col = ['ItemNumber',
            'ItemDescription',
            'Category',
            'CategoryName',
            'VendorNumber',
            'Date']

product = sales[product_col].copy()
```

## ðŸ§¹ Data Cleaning - Product Table

```python
# Rename some columns
product.rename(columns = {'ItemNumber':'ProductID',
                          'ItemDescription':'ProductDescription',
                          'Category':'CategoryCode',
                          'VendorNumber':'VendorID'}, inplace = True)
# Make lower-case
product['CategoryName'] = product['CategoryName'].str.lower()
product['ProductDescription'] = product['ProductDescription'].str.lower()
```

### I. Looking at ProductID numbers with missing CategoryCode data

```python
# Converting all of these to strings, some are numeric, others are strings
# cannot convert to numeric because I get an error for a combo ID, e.g. x1000 (not a real ID)
product['ProductID'] = product['ProductID'].apply(str)
```

```python
# Which ProductID numbers are missing CategoryCode data?
ProductIDs_missing_category = product.loc[product['CategoryCode'].isnull(),'ProductID'].unique()
ProductIDs_missing_category
```

```python
%%time
# if CategoryCode information is available on other rows, use the last CategoryCode & Category Name to fill in the information
for ID in ProductIDs_missing_category:
    # print(ID)
    temp = product.loc[product['ProductID'].isin([ID])] # create a smaller table of this product id
    temp = temp.sort_values(['Date'])
    ncodes_with_data = temp['CategoryCode'].nunique() 
    # print(ncodes_with_data)
    temp_notnull = temp.loc[temp['CategoryCode'].notnull()]

    # print(temp_notnull)
    if ncodes_with_data > 0:
        product.loc[((product['ProductID'].isin([ID]))&(product['CategoryCode'].isnull())),'CategoryName'] = temp_notnull['CategoryName'][temp_notnull.index[-1]]
        product.loc[((product['ProductID'].isin([ID]))&(product['CategoryCode'].isnull())),'CategoryCode'] = temp_notnull['CategoryCode'][temp_notnull.index[-1]]
```

```python
# The remaining ProductID numbers that are still missing CategoryCode data
product.loc[product['CategoryCode'].isnull(),'ProductID'].unique()
```
