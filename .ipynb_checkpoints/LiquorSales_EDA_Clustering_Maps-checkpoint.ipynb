# Store Analysis

```python
# Connect SQLite.
dbconn = sqlite3.connect('IowaLiquorSales.db')
cursor = dbconn.cursor()
```

```python
# Pull Store database.
query = '''SELECT * FROM Store;'''
cursor.execute(query)
result = cursor.fetchall()
cursor.close()
dbconn.close()
```

```python
# Put the database in a data frame format.
store = pd.DataFrame(result, columns = ['StoreNumber',
                                            'StoreName',
                                            'Address',
                                            'City',
                                            'ZipCode',
                                            'StoreLocation',
                                            'CountyNumber',
                                            'County'])
```

```python
# Connect SQLite.
dbconn = sqlite3.connect('IowaLiquorSales.db')
cursor = dbconn.cursor()
```

```python
# Pull Transactions database.
query = '''SELECT * FROM Transactions;'''
cursor.execute(query)
result = cursor.fetchall()
cursor.close()
dbconn.close()
```

```python
# Put the database in a data frame format.
transactions = pd.DataFrame(result, columns = ['ItemNumber',
                                               'Date',
                                               'StoreNumber', 
                                               'ProductNumber', 
                                               'VendorNumber', 
                                               'Category', 
                                               'BottlesSold', 
                                               'VolumeSold', 
                                               'Sale'])
```

```python
transactions.head()
```

```python
temp = transactions['Date']
temp_split0 = temp.str.split('/', n=- 1, expand=True)
temp_split = temp_split0.drop(temp_split0.columns[[1]], axis=1)

transactions["Year"] = temp_split[2]
```
