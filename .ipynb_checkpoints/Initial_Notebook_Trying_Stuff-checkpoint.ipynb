Create New Columns for Date

```python
temp = df['Date']
temp_split0 = temp.str.split('/', n=- 1, expand=True)
temp_split = temp_split0.drop(temp_split0.columns[[1]], axis=1)

df["Year"] = temp_split[2]
```

```python
df["Month"] = temp_split[0]

# coverting dates to datetime
ex = pd.to_datetime(df['Date'], format="%m/%d/%Y")

df["Weekday"] = ex.dt.weekday
```

```python
df.head(5)
```

```python
df.columns
```

# Location

```python
df.columns
```

```python
df['County'] = df['County'].str.upper()
```

```python
df['County'].unique()
```

```python
df_recent = df.loc[df['Year'].isin(["2016","2017","2018","2019","2020"])]
```

```python
df_recent.head()
```

```python
df_recent.groupby('Vendor Name').size().nlargest(20).sort_values(
    ascending=False)
```

```python
fig = plt.figure(figsize = (12,6))
df_recent.groupby('County')['Store Number'].size().nlargest(20).sort_values(
    ascending=False).plot.bar(color = "blue")

ax = plt.axes() 
ax.tick_params(axis='x', colors='black',rotation=45)
plt.ylabel('Number of Stores')
```

```python
fig = plt.figure(figsize = (12,6))
df_recent.groupby('County')['Volume Sold (Liters)'].sum().nlargest(20).sort_values(
    ascending=False).plot.bar(color = "blue")

ax = plt.axes() 
ax.tick_params(axis='x', colors='black',rotation=45)
plt.ylabel('Total Volume Sold (Liters)')
```
