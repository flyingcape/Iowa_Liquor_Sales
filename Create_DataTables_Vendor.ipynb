# Clean & Create Vendor Table

```python
vendor_col = ['VendorNumber',
              'VendorName',
              'Date']

vendor = sales[vendor_col].copy()
```

## 🧹 Data Cleaning - Vendor Table 

```python
# Rename some columns
vendor.rename(columns = {'VendorNumber':'VendorID'}, inplace = True)
```

```python
vendor.info()
```

```python
vendor['Date'] = pd.to_datetime(vendor['Date'])
```

```python
vendor.head()
```

```python
# Create a VendorID list with the most recent VendorName
VendorID_df = vendor.loc[(vendor['VendorName'].notnull())].sort_values(['VendorID','VendorName','Date']).drop_duplicates(subset = ['VendorID'], keep = 'last')
VendorID_df.head(25)
```

```python
# Consolidate the VendorName entries
Merged_Vendor_Names = vendor.merge(VendorID_df[['VendorID', 'VendorName']], on = 'VendorID', how = 'left')
Merged_Vendor_Names.drop(columns = 'VendorName_x', inplace = True)
Merged_Vendor_Names.rename(columns = {'VendorName_y': 'VendorName'}, inplace = True)
vendor = Merged_Vendor_Names.copy()
vendor = vendor.drop('Date',axis=1)
vendor.head()
```
